```python
import pandas as pd
import numpy as np
```

```python
data = []

def add(fir,las):
    temp = pd.read_csv('Dataset/metadata/ExtractedData_ReCovery_' + str(fir) + '_' + str(las) + '.csv',encoding='latin1')
    data.append(temp)
```

```python
add(0,1000)
add(1000,1400)
add(1400,2029)
```

```python
Merged = pd.concat(data, ignore_index=True, sort=False)
```

```python
Merged.head()
```

```python
Merged = Merged.drop(columns = ['Unnamed: 0'])
```
