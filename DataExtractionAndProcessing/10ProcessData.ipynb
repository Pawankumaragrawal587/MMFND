```python
source = pd.read_csv('Dataset/Initial_Data/Source/TICNN.csv')

source
```

```python
similarity = pd.read_csv('Dataset/Initial_Data/ImageSimilarities_TICNN.csv')

similarity
```

```python
similarity = similarity.iloc[:,:].values
```

```python
for i in range(len(similarity)):
    
    for j in range(2,6):
        
        similarity[i][j] = literal_eval(similarity[i][j])
        
        name = 'Image_url' + str(j-1)
        
        source[name][i] = literal_eval(source[name][i])
        
        temp = {}
        
        if len(similarity[i][j])>0:
            temp['image_name'] = source['ID'][i] + '_' + str(j-1) + '_' + str(similarity[i][j][0]['Index']) + '.jpg'
            temp['image_url'] = source[name][i][similarity[i][j][0]['Index']]
            source[name][i] = temp
        else:
            source[name][i] = []
```

```python
source
```

```python
source = source.drop(columns = ['Unnamed: 0'])

source
```

```python
source.to_csv('Dataset/Data/Source/TICNN.csv')
```
